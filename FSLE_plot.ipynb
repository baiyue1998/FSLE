```python
import numpy as np
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "fsle_20231010.nc"
ds = xr.open_dataset(file)
```

```python
lat = ds['lat']
lon = ds['lon']

lambda1 = ds['lambda1']
```

```python
figure1 = plt.figure(figsize=(6,4.5))
plt.pcolormesh(lon,lat,lambda1,cmap = 'terrain',shading='nearest',vmax=0,vmin=-1)
plt.colorbar(extend='both')
plt.xlabel('Longitude [E]',fontsize=14)
plt.ylabel('Latitude [N]',fontsize=14)
```
